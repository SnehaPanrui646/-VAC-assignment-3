```python
df=np.arange(10,15)
df1=np.array([11,12,13,14,15])
print(df)
print(df1)
np.corrcoef(df,df1)
```

```python

reg=scipy.stats.linregress(df,df1)
print(reg)
print(reg.slope)
print(reg.rvalue)
print(reg.pvalue)
print(reg.stderr)
print(reg.intercept)
print(reg.intercept_stderr)
```
