```python
df=np.arange(8,13)
df1=np.array([9,10,11,12,13])
print(df)
print(df1)
np.corrcoef(df,df1)
```

```python
df=np.arange(15,20)
df1=np.array([10,14,18,22,26])
r=np.corrcoef(df,df1)
r1=scipy.stats.pearsonr(df,df1)
print(r1[0],r1[1])
r2=scipy.stats.spearmanr(df,df1)
print(r2[0],r2[1])
r3=scipy.stats.kendalltau(df,df1)
print([r3[0],r3[1]])
```

```python
reg=scipy.stats.linregress(df,df1)
print(reg)
print(reg.slope)
print(reg.rvalue)
print(reg.pvalue)
print(reg.stderr)
print(reg.intercept)
print(reg.intercept_stderr)
```
